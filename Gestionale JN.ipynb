```python
import csv 
from datetime import datetime

file_name='transazioni.csv'
header=['DATA','causale','importo speso']
with open(file_name, mode="w") as file:
      writer = csv.writer(file)
      writer.writerow(header)
      


#ciclo while con cui posso tornare al menu una volta svolte le operazioni in uno dei 3 blocchi if
while True:                      
      menu= input('\nciao questo è il menu, scegli cosa vuoi fare:\n \
            \naggiungi operazione >seleziona 1\n\
            \nreport mensile >seleziona2\n\
            \ntop 10 transazioni >seleziona3\n>')


#menu 1 (aggiunta di operazione):
#splitta l'input in DATA, CAUSALE, IMPORTO per poterlo salvare correttamente nel csv
      if menu==   '1':
            date_format='%d/%m/%Y'
      
            while True:
                  transazione = input('\n inserisci data, descrizione e importo che vuoi aggiungere \n \
                  (formato: GG/MM/YYYY causale importo)')
                  parts = transazione.split()
                  importo = parts[-1]
                  causale = ' '.join(parts[1:-1])
                  data = parts[0]

#controlla che la data inserita è del formato giusto (break while), altrimenti torna all'inserimento dell'operazione
                  try:
                        date =datetime.strptime(data, date_format)
                        break
                  except ValueError:
                        print("\nFORMATO NON CORRETTO !")
 #salvataggio nel csv di DATA, CAUSALE, IMPORTO separati da virgola        
            with open (r'transazioni.csv',"a") as file:
                  nuova_riga= '\n'+ data +','+ causale +','+ importo
                  writer = file.write(nuova_riga)
                  print('operazione aggiunta con successo')
      

#menu 2 (report diviso per mese e anni):
# legge il csv e crea una lista di dizionari da cui poter estrarre le voci: data(anno e mese) e importo da sommare      
      
      elif menu =='2':
            sum_by_month_year={}
            with open('transazioni.csv') as csvfile:
                  reader = csv.DictReader(csvfile)      
                  rows=list(reader)# creo una lista di dizionari(righe) per poter accedere facilmente
            
            for row in rows:
            
                  month_year = row['DATA'][3:10]   #MM/YYYY
                  importo_speso = float(row['importo speso'])

#ciclo if che aggiunge al dizionario vuoto sum_by_month_year:
#la data e l importo la prima volta; poi quando trova la data uguale somma l'importo
#infine ordina tutto il dizionario per data con la lambda function e strptime
                  if month_year in sum_by_month_year:
                        sum_by_month_year[month_year] += importo_speso
                  else:
                        sum_by_month_year[month_year] = importo_speso
            sorted_dates = sorted(sum_by_month_year.keys(), key=lambda x: datetime.strptime(x, '%m/%Y'))


            print('\n----------REPORT SPESE MESILI----------\n')

#ciclo for per riformattare la data da MM/YYYY  a   YYYY-MM
# viene introdotto un conteggio (previous e current year) in modo che quando inizia un anno diverso lascia una riga vuota            
            previous_year = None
            
            for month_year in sorted_dates:
                  
                  current_year=month_year[3:]
                  new_format = f"{month_year[3:]}-{month_year[:2]}"   #inverte MM/YYYY con YYYY-MM   

# se l'anno corrente è diverso da quello prima allora stampa una riga vuota prima di iniziare l'anno nuovo                
                  if current_year != previous_year and previous_year is not None:
                        print()           
                  print(f"Anno/Mese: {new_format}, Totale Importo Speso: € {int(sum_by_month_year[month_year])}")
                  
                  previous_year= current_year  #il conteggio si sposta in avanti
            
            
            print('\n' + '=' * 40)
            print('        FINE REPORT SPESE MENSILI')
            print('========================================')
            print(' +   +   +   +   +   +   +   +   +   +   ')
            print('   +   +   +   +   +   +   +   +   +   +  ')

            input("\nPremi INVIO per tornare al menu principale...")
      
      
      
      
      
#menu 3 (top 10 operazioni):
       
      elif menu =='3':
            with open('transazioni.csv') as csvfile:
                  reader = csv.DictReader(csvfile)
                  rows = list(reader)

# Ordina le righe della lista per importo (converto l'importo in float)
            rows_sorted = sorted(rows, key=lambda row: float(row['importo speso']))

# Stampa solo le prime 10 righe ordinate per importo
            for index, row in enumerate(rows_sorted[-10:]):
                  print( row['DATA']+' '+row['causale']+' '+row['importo speso'])

            input("\nPremi INVIO per tornare al menu principale...")
#premuto invio torna al ciclo while iniziale (menu iniziale)
```
